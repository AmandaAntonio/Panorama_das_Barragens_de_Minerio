Barragem de rejeitos  um
reservatrio destinado a reter resduos
resultantes 
de processos de extrao de
minrios. O rejeito, resultado dessa
extrao,  um material que no
possui valor comercial, mas para fins de preservaes ambientais deve ser
devidamente armazenado. 

No Brasil, existe a Poltica Nacional
de Segurana de Barragens 
PNSB, criada pela Lei n.
12.334/2010, que tem por objetivo
garantir padres de segurana de
barragens na perspectiva de
prevenir ou reduzir a possibilidade
de acidentes ou desastres e suas
consequncias. Essa lei atribui ao empreendedor
(responsvel pela barragem) o
dever de cuidar e manter a
barragem segura, e ao poder
pblico o dever de fiscaliz-las.

Mesmo com a existncia de leis regulamentadoras e padres de seguranas estabelecidos, em 25 de janeiro de 2019 no municpio de Brumadinho-MG a barragem da Mina Crrego do Feijo se rompeu causando a morte de 270 pessoas, incluindo quatro desaparecidas, e um dos maiores desastres ambientais do pas.

As barragem recebem classificao: Alta, Mdia e Baixa.

Dano Potencial Associado -  o dano que pode ocorrer devido a rompimento, vazamento, infiltrao no solo ou mau funcionamento de uma barragem, independentemente da sua probabilidade de ocorrncia, podendo ser graduado de acordo com as perdas de vidas humanas e impactos sociais, econmicos e ambientais. As barragem recebem classificao: Alta, Mdia e Baixa.

Classe -  uma matriz de notas de A a E, dada pelas classificaes de categoria de risco e dano potencial.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Carregando os Dados das Barragens


```python
df = pd.read_csv('/content/brasil_barragens.csv')
```

```python
df.head()
```

# Conhecendo o Dataframe



```python
# Verificando as informaes do dataframe
df.info()
```

```python
#Verificando os dados da coluna de volume
df.VOLUME_ATUAL_m3.sort_values()
```

```python
#Verificando o dados da coluna Altura
df.ALTURA_ATUAL_metros.sort_values()
```

```python
# Ajustando e convertendo os valores de altura e volume 
df['ALTURA_ATUAL_metros'] = df['ALTURA_ATUAL_metros'].astype(str).str.replace(',', '')#trocando virgula por espao vazio
df['ALTURA_ATUAL_metros'] = df['ALTURA_ATUAL_metros'].replace('-', np.nan)#trocando hifen por nulos(vieram hifen como nulos)
df['ALTURA_ATUAL_metros'] = pd.to_numeric(df['ALTURA_ATUAL_metros'],errors='coerce')#convertendo para dados numerico

df['VOLUME_ATUAL_m3'] = df['VOLUME_ATUAL_m3'].astype(str).str.replace(',','')#trocando virgula por espao vazio
df['VOLUME_ATUAL_m3'] = df['VOLUME_ATUAL_m3']. replace('-', np.nan)#trocando hifen por nulos(vieram hifen como nulos)
df['VOLUME_ATUAL_m3'] = pd.to_numeric(df['VOLUME_ATUAL_m3'],errors='coerce')#convertendo para dados numerico
```

```python
# Verificando o resltado
df.info()
```

# Explorando os Dados

```python
#Porcentagem de valores missing
na_cols = [col for col in df if (df[col].isna().sum() > 0)] #selecionando colunas com valores missing

#Criando o grfico
ax = (
    df[na_cols]
    .isna()
    .mean() # trazer o percentual de missing
    .sort_values(ascending=False) #ordena de forma decrescente
    .mul(100) #convertendo para centenas
    .plot(kind='barh', figsize=(12,6)) # grafico na horizontal, tamanho
)

_= ax.set_xlabel('Porcentagem de Missing') # titulo do eixo x
_=ax.set_title('Anlise de Missing') #titulo principal
```

```python
#Selecionandos od dados missing
df[df['CATEGORIA_DE_RISCO'].isna()].head()#selecionando uma amostra dos dados
```

Chama ateno o fato de que a coluna categoria de risco no apresenta 40% de dados e a principio no encontram-se enquadradas na PSNB. Dentre as possveis causas pensadas destacam-se:

No site [ANM](https://app.anm.gov.br/SIGBM/Publico/ClassificacaoNacionalDaBarragem) a possvel fonte de origem, foi possivel verificar que:


1.   Existem dados missing na base fornecida no site
2.   Existem critrios para que uma barragem precise se enquadrar a PSNB.

```python
#Comparando os critrios e os dados
df[df['CATEGORIA_DE_RISCO'].isna()].describe()#selecionando os dados e vericando o resumo estatstico
```

```python
#Barragem com o valores mximo na categoria de risco
df[df['CATEGORIA_DE_RISCO'].isna()].max()
```

```python
#Criando dummy para avaliar as variveis categricas da coluna PSNB
df['dummy_PNSB'] = np.where(df['INSERIDA_NA_PNSB'] == "Sim", 1, 0)#selecionando a coluna e determinando o criterio
df['dummy_PNSB'].value_counts()#contando os valores por criterio
```

```python
#Percentual de barragens englobadas na PNSB
df['dummy_PNSB'].value_counts(normalize=True)#valor percentual do dummy
```

Os dados das barragens contendo missing respeitam os critrios da PSNB e por isso no se encaixam na avaliao de risco.



```python
#Resumo dos dados por tipo de minrio
resumo = (
    df.groupby(['MINERIO_PRINCIPAL'], sort=False) #agrupar por tipo de minrio
    .agg( #agregar os dados
        sum_pnsb=('dummy_PNSB', 'sum'), #qtd de barragens esto na PSNB por tipo de minrio
        max_altura=('ALTURA_ATUAL_metros', 'max'),#max de altura por tipo de minrio
        max_volume=('VOLUME_ATUAL_m3', 'max'),#max de volume por tipo de minrio
        media_altura=('ALTURA_ATUAL_metros', 'mean'), #mdia altura por tipo de minrio
        media_volume=('VOLUME_ATUAL_m3', 'mean'), #mdia volume por tipo de minrio
        sum_barragem=('MINERIO_PRINCIPAL', 'count') #qtd de barragens por tipo de minrio
        ).reset_index()
)

resumo.sort_values(by='sum_pnsb').head()#vericando uma amostra dos dados e ordenando os valores
```

```python
#Verificando o resumo 
resumo.sort_values(by='sum_pnsb')
```

# Qual Estado possui maior porcentagem de barragens com dano potencial alto?

```python
#Criando dummys para dano potencial
df = pd.concat( #concatenando 
    [
        df, #fonte de dados
        pd.get_dummies(df['DANO_POTENCIAL_ASSOCIADO'], 'Dano_Potencial'), #criando as dummys
    ],
    axis=1, #anexar ao eixo das colunas
)
```

```python
#Conferindo o resultado
df.head(2)
```

```python
#Grfico da porcentagem de barragens com dano potencial alto
_ = (
    pd.pivot_table( #criando uma tabela com os dados de dano potencial alto
        data=df, index=['UF'], values=['Dano_Potencial_Alta'], aggfunc='mean') #indicando o dataframe e os valores
    .sort_values(by='Dano_Potencial_Alta', ascending=False) #ordenao
    .plot(kind='bar') #tipo de plot
)
_ = plt.title("Porcentagem de Barragens com Dano Potencial Alto") #titulo do grfico
```

# Verificando a distribuio por Estado de barragens com dano potencial alto e categoria de risco alta

```python
#Criando dummy para categoria de risco
df = pd.concat(
    [
        df,
        pd.get_dummies(df['CATEGORIA_DE_RISCO'], prefix='Categoria_de_Risco'),
    ],
    axis=1,
)
```

```python
#Criando um resumo por UF
resumo_por_uf = (
       df.groupby(['UF'], sort=False) #agrupando cada estado
      .agg( #agragar os dados
        contagem_geral=('Dano_Potencial_Alta', 'count'), #qtd. de barragens
        contagem_por_risco_alto=('Categoria_de_Risco_Alta', 'sum'), #qtd. de barragens com alto risco
        perc_de_risco_alto=('Categoria_de_Risco_Alta', 'mean'), #percentual de barragens com alto risco
        contagem_por_dano_alto=('Dano_Potencial_Alta', 'sum'), #qtd. de barragens com dano alto
        perc_de_dano_alto=('Dano_Potencial_Alta', 'mean'), #percentual de barragens com dano alto
    )
    .reset_index() #resetando o index
    .round(1) #arrendondando os valores
)
```

```python
#verificando o resulatado
resumo_por_uf.sort_values(by='contagem_geral', ascending=False)
```
