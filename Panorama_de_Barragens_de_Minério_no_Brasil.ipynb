# Barragens de Rejeitos no Brasil


barragem_minerio.jpeg

Barragem de rejeitos é um
reservatório destinado a reter resíduos
resultantes 
de processos de extração de
minérios. O rejeito, resultado dessa
extração, é um material que não
possui valor comercial, mas para fins de preservações ambientais deve ser
devidamente armazenado. 

No Brasil, existe a Política Nacional
de Segurança de Barragens –
PNSB, criada pela Lei n.º
12.334/2010, que tem por objetivo
garantir padrões de segurança de
barragens na perspectiva de
prevenir ou reduzir a possibilidade
de acidentes ou desastres e suas
consequências. Essa lei atribui ao empreendedor
(responsável pela barragem) o
dever de cuidar e manter a
barragem segura, e ao poder
público o dever de fiscalizá-las.

Mesmo com a existência de leis regulamentadoras e padrões de seguranças estabelecidos, em 25 de janeiro de 2019 no município de Brumadinho-MG a barragem da Mina Córrego do Feijão se rompeu causando a morte de 270 pessoas, incluindo quatro desaparecidas, e um dos maiores desastres ambientais do país.

Afim de aumentar o conhecimento e a conscientização sobre o assunto, desenvolvi está análise de dados.

As barragem recebem classificação: Alta, Média e Baixa.

Dano Potencial Associado - é o dano que pode ocorrer devido a rompimento, vazamento, infiltração no solo ou mau funcionamento de uma barragem, independentemente da sua probabilidade de ocorrência, podendo ser graduado de acordo com as perdas de vidas humanas e impactos sociais, econômicos e ambientais. As barragem recebem classificação: Alta, Média e Baixa.

Classe - é uma matriz de notas de A a E, dada pelas classificações de categoria de risco e dano potencial.

# Carregando as Bibliotecas




```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Carregando os Dados das Barragens


```python
df = pd.read_csv('/content/brasil_barragens.csv')
```

```python
df.head()
```

# Conhecendo o Dataframe



```python
# Verificando as informações do dataframe
df.info()
```

```python
#Verificando os dados da coluna de volume
df.VOLUME_ATUAL_m3.sort_values()
```

```python
#Verificando o dados da coluna Altura
df.ALTURA_ATUAL_metros.sort_values()
```

```python
# Ajustando e convertendo os valores de altura e volume 
df['ALTURA_ATUAL_metros'] = df['ALTURA_ATUAL_metros'].astype(str).str.replace(',', '')#trocando virgula por espaço vazio
df['ALTURA_ATUAL_metros'] = df['ALTURA_ATUAL_metros'].replace('-', np.nan)#trocando hifen por nulos(vieram hifen como nulos)
df['ALTURA_ATUAL_metros'] = pd.to_numeric(df['ALTURA_ATUAL_metros'],errors='coerce')#convertendo para dados numerico

df['VOLUME_ATUAL_m3'] = df['VOLUME_ATUAL_m3'].astype(str).str.replace(',','')#trocando virgula por espaço vazio
df['VOLUME_ATUAL_m3'] = df['VOLUME_ATUAL_m3']. replace('-', np.nan)#trocando hifen por nulos(vieram hifen como nulos)
df['VOLUME_ATUAL_m3'] = pd.to_numeric(df['VOLUME_ATUAL_m3'],errors='coerce')#convertendo para dados numerico
```

```python
# Verificando o resltado
df.info()
```

# Explorando os Dados

```python
#Porcentagem de valores missing
na_cols = [col for col in df if (df[col].isna().sum() > 0)] #selecionando colunas com valores missing

#Criando o gráfico
ax = (
    df[na_cols]
    .isna()
    .mean() # trazer o percentual de missing
    .sort_values(ascending=False) #ordena de forma decrescente
    .mul(100) #convertendo para centenas
    .plot(kind='barh', figsize=(12,6)) # grafico na horizontal, tamanho
)

_= ax.set_xlabel('Porcentagem de Missing') # titulo do eixo x
_=ax.set_title('Análise de Missing') #titulo principal
```

```python
#Selecionandos od dados missing
df[df['CATEGORIA_DE_RISCO'].isna()].head()#selecionando uma amostra dos dados
```

```python
#Comparando os critérios e os dados
df[df['CATEGORIA_DE_RISCO'].isna()].describe()#selecionando os dados e vericando o resumo estatístico
```

```python
#Barragem com o valores máximo na categoria de risco
df[df['CATEGORIA_DE_RISCO'].isna()].max()
```

```python
#Criando dummy para avaliar as variáveis categóricas da coluna PSNB
df['dummy_PNSB'] = np.where(df['INSERIDA_NA_PNSB'] == "Sim", 1, 0)#selecionando a coluna e determinando o criterio
df['dummy_PNSB'].value_counts()#contando os valores por criterio
```

```python
#Percentual de barragens englobadas na PNSB
df['dummy_PNSB'].value_counts(normalize=True)#valor percentual do dummy
```

Os dados das barragens contendo missing respeitam os critérios da PSNB e por isso não se encaixam na avaliação de risco.



```python
#Resumo dos dados por tipo de minério
resumo = (
    df.groupby(['MINERIO_PRINCIPAL'], sort=False) #agrupar por tipo de minério
    .agg( #agregar os dados
        sum_pnsb=('dummy_PNSB', 'sum'), #qtd de barragens estão na PSNB por tipo de minério
        max_altura=('ALTURA_ATUAL_metros', 'max'),#max de altura por tipo de minério
        max_volume=('VOLUME_ATUAL_m3', 'max'),#max de volume por tipo de minério
        media_altura=('ALTURA_ATUAL_metros', 'mean'), #média altura por tipo de minério
        media_volume=('VOLUME_ATUAL_m3', 'mean'), #média volume por tipo de minério
        sum_barragem=('MINERIO_PRINCIPAL', 'count') #qtd de barragens por tipo de minério
        ).reset_index()
)

resumo.sort_values(by='sum_pnsb').head()#vericando uma amostra dos dados e ordenando os valores
```

```python
#Verificando o resumo 
resumo.sort_values(by='sum_pnsb')
```

# Qual Estado possui maior porcentagem de barragens com dano potencial alto?

```python
#Criando dummys para dano potencial
df = pd.concat( #concatenando 
    [
        df, #fonte de dados
        pd.get_dummies(df['DANO_POTENCIAL_ASSOCIADO'], 'Dano_Potencial'), #criando as dummys
    ],
    axis=1, #anexar ao eixo das colunas
)
```

```python
#Conferindo o resultado
df.head(2)
```

```python
#Gráfico da porcentagem de barragens com dano potencial alto
_ = (
    pd.pivot_table( #criando uma tabela com os dados de dano potencial alto
        data=df, index=['UF'], values=['Dano_Potencial_Alta'], aggfunc='mean') #indicando o dataframe e os valores
    .sort_values(by='Dano_Potencial_Alta', ascending=False) #ordenação
    .plot(kind='bar') #tipo de plot
)
_ = plt.title("Porcentagem de Barragens com Dano Potencial Alto") #titulo do gráfico
```

# Verificando a distribuição por Estado de barragens com dano potencial alto e categoria de risco alta

```python
#Criando dummy para categoria de risco
df = pd.concat(
    [
        df,
        pd.get_dummies(df['CATEGORIA_DE_RISCO'], prefix='Categoria_de_Risco'),
    ],
    axis=1,
)
```

```python
#Criando um resumo por UF
resumo_por_uf = (
       df.groupby(['UF'], sort=False) #agrupando cada estado
      .agg( #agragar os dados
        contagem_geral=('Dano_Potencial_Alta', 'count'), #qtd. de barragens
        contagem_por_risco_alto=('Categoria_de_Risco_Alta', 'sum'), #qtd. de barragens com alto risco
        perc_de_risco_alto=('Categoria_de_Risco_Alta', 'mean'), #percentual de barragens com alto risco
        contagem_por_dano_alto=('Dano_Potencial_Alta', 'sum'), #qtd. de barragens com dano alto
        perc_de_dano_alto=('Dano_Potencial_Alta', 'mean'), #percentual de barragens com dano alto
    )
    .reset_index() #resetando o index
    .round(1) #arrendondando os valores
)
```

```python
#verificando o resulatado
resumo_por_uf.sort_values(by='contagem_geral', ascending=False)
```
